```python
null_precip = df[df['Precip Type'].isnull()]
print(null_precip[['Temperature (C)', 'Humidity', 'Visibility (km)']].describe())
```

REPLACING THE NULL VALUES

```python

df.loc[(df['Precip Type'].isnull()) &
       (df['Humidity'] > 0.7), 'Precip Type'] = 'rain'

df.loc[(df['Precip Type'].isnull()) &
       (df['Temperature (C)'] < 0), 'Precip Type'] = 'snow'
df['Precip Type'] = df['Precip Type'].fillna(df['Precip Type'].mode()[0])
```

```python
x=df.drop(['Humidity','Formatted Date','Summary','Precip Type','Daily Summary'],axis=1)
```

```python
y=df['Humidity']
```

# DATA MODELING

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
reg_predict=lr.predict(x_test)
```

```python
reg_predict
```

```python
sns.distplot(reg_predict-y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, reg_predict)

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, reg_predict)

# R-squared (R²)
r2 = r2_score(y_test, reg_predict)

print(f"MAE: {mae:.2f}, MSE: {mse:.2f}, R²: {r2:.2f}")
```

```python
from sklearn.linear_model import Ridge
rid=Ridge()
```

```python
rid.fit(x_test,y_test)
```

```python
pred=rid.predict(x_test)
```

```python
sns.distplot(pred-y_test)
```

```python

# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, pred)

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, pred)

# R-squared (R²)
r2 = r2_score(y_test, pred)

print(f"MAE: {mae:.2f}, MSE: {mse:.2f}, R²: {r2:.2f}")
```
